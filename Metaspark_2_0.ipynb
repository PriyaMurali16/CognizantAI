```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
#Loading the datasets and using the describe and info fuction on them
data_1 = pd.read_csv("/content/drive/MyDrive/metaspark/log_action.csv")
data_2 = pd.read_csv("/content/drive/MyDrive/metaspark/log_link_visit_action.csv")
data_3=pd.read_csv("/content/drive/MyDrive/metaspark/log_visit.csv")
```

```python
#Exploring the columns in the datasets
print (data_1.columns)
print (data_2.columns)
print (data_3.columns)
```

```python
# Columns that have no unique values , do not contribute to our prediction dependent variable
for column_name in data_2.columns:
  if data_2[column_name].nunique() == 0:
     data_2.drop(column_name,inplace=True,axis=1)
```

```python
for column_name in data_3.columns:
  if data_3[column_name].nunique() == 0:
     data_3.drop(column_name,inplace=True,axis=1)
```

```python
X=data_merged.drop('visit_goal_converted',axis=1)
```

```python
Y=data_merged['visit_goal_converted']
```

```python
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size,
random_state=seed)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
```

```python
Y_pred = logreg.predict(X_test)
```

```python

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, Y_pred)
print ("Confusion Matrix : \n", cm)
```

```python
from sklearn.metrics import accuracy_score

print ("Accuracy : ", accuracy_score(Y_test, Y_pred))
```
